```python
from pyspark.sql.functions import col

df_clean = complex_joined_df \
    .withColumnRenamed("Invoice No.", "Invoice_No") \
    .withColumnRenamed("Stock code", "Stock_Code") \
    .withColumnRenamed("Invoice Date", "Invoice_Date") \
    .withColumnRenamed("Unit Price", "Unit_Price") \
    .withColumnRenamed("Customer ID", "Customer_ID")
```

```python
df_clean = df_clean.withColumn(
    "Total_Amount",
    col("Quantity") * col("Unit_Price")
)

display(df_clean)
```
